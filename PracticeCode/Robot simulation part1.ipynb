```python
import numpy as np
import matplotlib.pyplot as plt
import math
import csv
#"step" function by robot

#def func(x,U):
    #-2x+2U
    #val = -(2*x)+2*U
    #return val

#different order of parameters passed different behaviour??????
def func(U,x):
    #-2x+2U
    val = -(2*x) + 2*U
    return val
```

```python
a = 0
b = 0
#append every 10th step if h value less than or equal to 0.01
t = a
x = b
#h =1 #step size
h = 0.01

##i want to save every 10 samples
save = 0
U_save = []
t_save = []
x_save = []
x_exact_save = []


##saves every time
U_array = []
t_array = []
x_array = []
x_exact_array = []
while(t <= 15):
    if(t <= 5):
      U = 2
    elif(t < 5 or t <= 10):
      U = 1
    elif(t < 10 or t <=15):
      U = 3
    print("U value:", x)
    U_array.append(U)
    k = h * func(U,x)
    x = x + k
    t = t + h
    #value under graph
    x_exact = U - math.exp(-2*t)
    t_array.append(t)
    x_array.append(x)
    x_exact_array.append(x_exact)
    save+=1
    print("save: ",save)
    if(save%10 == 0):
        U_save.append(U)
        t_save.append(t)
        x_save.append(x)
        x_exact_save.append(x_exact)
        print("save1: ",save)
    print("Where it is: ",x,"Exact pos= ",x_exact,"Time: ",t)
    #plt.fill_between(x_axis,y_axis, where = [(x>0)and (x <2)for x_axis in x_axis], color = 'green')
```
